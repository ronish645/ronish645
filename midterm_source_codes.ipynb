```python
#bubble sorting example
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        # Track if any swap happens
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        # If no two elements were swapped, the list is sorted
        if not swapped:
            break
    return arr

# Example usage:
arr = [2, 5, 7, 1, 9, 6]
print("Initial List:", arr)
sorted_arr = bubble_sort(arr)
print("Sorted List:", sorted_arr)
```

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        # Find the minimum element in remaining unsorted array
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        # Swap the found minimum element with the first element
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

# Example usage:
arr = [2, 5, 7, 1, 9, 6]
print("Initial List:", arr)
sorted_arr = selection_sort(arr)
print("Sorted List:", sorted_arr)
```

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# Example usage:
arr = [2, 5, 7, 1, 9, 6]
print("Initial List:", arr)
sorted_arr = insertion_sort(arr)
print("Sorted List:", sorted_arr)
```

```python
def count_letter(n):
    # Initialize an empty dictionary to store letter counts
    letter_counts = {}

    for letter in n:
        # Process only letters
        if letter.isalpha():
            # Convert letter to lowercase
            letter = letter.lower()
            # Update the count in the dictionary
            if letter in letter_counts:
                letter_counts[letter] += 1
            else:
                letter_counts[letter] = 1

    # Display the letter counts
    for letter, count in letter_counts.items():
        print(f"{letter}: {count}")

# Prompt the user to enter a string
n = "Hello world!"
# Call the function to count and display letter occurrences
count_letter(n)
```

```python
# reverse the string using stacks
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return None

    def is_empty(self):
        return len(self.items) == 0

# Function to reverse a string using a stack
def reverse_string(input_string):
    stack = Stack()

    # Push all characters of the string onto the stack
    for char in input_string:
        stack.push(char)

    # Pop all characters from the stack and form the reversed string
    reversed_string = ''
    while not stack.is_empty():
        reversed_string += stack.pop()

    return reversed_string

# Given string
input_string = "Hello people Im here!"

# Reverse the string using the stack
reversed_string = reverse_string(input_string)
print("Original String:", input_string)
print("Reversed String:", reversed_string)
```

```python
# FIFO example
from collections import deque

# Initialize a queue
ticket_queue = deque()

# Enqueue people
ticket_queue.append('Person 1')
ticket_queue.append('Person 2')
ticket_queue.append('Person 3')

# Dequeue people as they are served
served_person1 = ticket_queue.popleft()
print("Served:", served_person1)  # Output: Served: Person 1

served_person2 = ticket_queue.popleft()
print("Served:", served_person2)  # Output: Served: Person 2

print("Remaining in queue:", list(ticket_queue))  # Output: Remaining in queue: ['Person 3']
```

```python
#LIFO example
# Initialize a stack
stack = []

# Push elements onto the stack
stack.append('Book 1')
stack.append('Book 2')
stack.append('Book 3')

print("Stack after pushing:", stack)

# Pop elements from the stack
last = stack.pop()
print("Popped:", last)

second_last = stack.pop()
print("Popped:", second_last)

print("Stack after popping:", stack)
```

```python
#14 priority queues example
import heapq

# Initialize an empty priority queue
priority_queue = []

# Add tasks to the priority queue with their priorities
heapq.heappush(priority_queue, (3, 'Pack luggage'))
heapq.heappush(priority_queue, (1, 'Book a taxi'))
heapq.heappush(priority_queue, (2, 'Check flight status'))
heapq.heappush(priority_queue, (4, 'Leave house'))

# Remove tasks from the priority queue based on their priorities
while priority_queue:
    priority, task = heapq.heappop(priority_queue)
    print(f"Priority: {priority}, Task: {task}")
```
